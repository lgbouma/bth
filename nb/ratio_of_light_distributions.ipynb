```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
max_len = 5e4

q_list, γ_R_list = [], []

while len(q_list)<max_len:
    q = np.random.uniform(low=0.1, high=1
    
    α = 2.6 if q < 0.5 else 4.5
    
    γ_R = q**α
    
    q_list.append(q)
    γ_R_list.append(γ_R)
```

```python
f,ax = plt.subplots()
hist, bin_edges = np.histogram(q_list, bins=np.arange(0,1.2,0.1), normed=True)
ax.step(bin_edges[:-1], hist, where='post')
ax.set(xlabel='q', ylabel='prob');
```

```python
f,ax = plt.subplots()
hist, bin_edges = np.histogram(γ_R_list, bins=np.append(np.logspace(-3, 0, 13), 42), normed=True)
ax.step(bin_edges[:-1], hist, where='post')
ax.set_xscale('log')
ax.set(xlabel='gamma_R', ylabel='prob');
```

```python
f,ax = plt.subplots()
hist, bin_edges = np.histogram(γ_R_list, bins=np.append(np.linspace(0,1,100),42), normed=True)
ax.step(bin_edges[:-1], hist, where='post')
ax.set(xlabel='L2/L1 ($\gamma_R$)',
       ylabel='prob',
       title='for realistic distribution, most binaries have small L2/L1')
f.savefig('gamma_R_distribution_numeric.pdf', dpi=300);
```

```python
import pandas as pd
df = pd.DataFrame(γ_R_list, columns=['gamma_R'])
df.describe()
```
